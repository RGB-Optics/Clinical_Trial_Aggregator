```python
import requests
import pandas as pd
```

```python
def fetch_clinical_trials(keywords, status="Recruiting|Not yet recruiting", study_type="Interventional", country="United States"):
    """
    Fetch clinical trial data from clinicaltrials.gov using the modern API.
    """
    # Base URL for clinicaltrials.gov API
    base_url = "https://clinicaltrials.gov/api/query/study_fields"
    
    # Fields to fetch from the API
    fields = [
        "NCTId", "BriefTitle", "Condition", "InterventionName", 
        "OverallStatus", "LocationCountry", "LocationCity", 
        "LocationState", "LocationFacility", "StudyType", "Phase"
    ]
    
    # API parameters
    params = {
        "expr": f"{keywords} AND {status} AND {study_type} AND {country}",
        "fields": ",".join(fields),
        "min_rnk": 1,
        "max_rnk": 1000,  # Number of results to fetch (can be adjusted)
        "fmt": "json"  # JSON format
    }
    
    # Make the API request
    response = requests.get(base_url, params=params)
    
#     if response.status_code != 200:
#         raise Exception(f"Failed to fetch data: {response.status_code}")
    
#     # Parse the JSON response
#     data = response.json()
#     trials = data.get("StudyFieldsResponse", {}).get("StudyFields", [])
    
#     if not trials:
#         print("No trials found.")
#         return None
    
#     # Convert to a DataFrame for easier processing
#     df = pd.DataFrame(trials)
    
#     # Flatten lists in DataFrame columns for better readability
#     for col in df.columns:
#         df[col] = df[col].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
    
    return params, response
```

```python
# Example usage
keywords = "Multiple System Atrophy"
status = "Recruiting|Not yet recruiting"
study_type = "Interventional"
country = "United States"
```

```python
fetch_clinical_trials(keywords, status, study_type, country)
```
